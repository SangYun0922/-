# 조달청 공공조달 빅데이터 경진대회
## 외자 계약 상세내역 데이터 시각화
### 등록유형 "자체전자조달시스템"일때는 원산지 기재되어 있지 않음
> 가장 많은 외자거래를 진행한 물품을 내림차순으로 정렬 및
> 금액별로 내림차순 정렬후, 근 3~5년간 가장 많은 지출을 한 항목을 통계치로 제시
>> 해당 물품에 대해서 국내 생산 권유
>>> 각 물품의 종류와 통화국가를 어떻게 시각화를 할지 고민중...

```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import rc  
rc('font', family='AppleGothic') 
plt.rcParams['axes.unicode_minus'] = False
```

```python
#등록유형 "나라장터(G2B)"일때, 통화국가 통계(2021.01.13~ 2022.01.13)
df = pd.read_csv("외자계약상세내역.csv")
df = df[df['등록유형'] == '나라장터(G2B)']
df['통화국가명'].value_counts().plot(kind='bar',figsize=(20,20), fontsize=15)
plt.title("나라장터 외자거래 통화국가", fontsize=20)
plt.xlabel("통화국가명", fontsize=20)
plt.ylabel("등록건수", fontsize=20)
```

```python
#등록유형 "나라장터(G2B)"일때, 원산지 통계(2021.01.13~ 2022.01.13)
df = pd.read_csv("외자계약상세내역.csv")
df = df[df['등록유형'] == '나라장터(G2B)']
df['원산지'].value_counts().plot(kind='bar',figsize=(20,20), fontsize=15)
plt.title("나라장터 외자거래 원산지", fontsize=20)
plt.xlabel("원산국가명", fontsize=20)
plt.ylabel("등록건수", fontsize=20)
```

```python
#등록유형 "나라장터(G2B)"일때, 물품명 통계(2021.01.13~ 2022.01.13)
df = pd.read_csv("외자계약상세내역.csv")
df = df[df['등록유형'] == '나라장터(G2B)']
df['품명'][:50].value_counts().plot(kind='bar',figsize=(20,20), fontsize=15)
plt.title("나라장터(G2B) 외자거래 물품명", fontsize=20)
plt.xlabel("품명", fontsize=20)
plt.ylabel("빈도수", fontsize=20)
```

```python
#등록유형 "나라장터(G2B)"일때, 물품명 통계(2021.01.13~ 2022.01.13)
df = pd.read_csv("외자계약상세내역.csv")
df = df[df['등록유형'] == '나라장터(G2B)']
df_2 = df.loc[:,['품명', '계약금액(USD)']]
grouped = df_2.groupby('품명')
df_sum = grouped.sum()
df_sum = df_sum.sort_values(by ='계약금액(USD)' ,ascending=False)
df_sum[:50].plot(kind='bar',figsize=(20,20), fontsize=15)
plt.title("나라장터 통화국가별 계약수량 통계 ", fontsize=20)
plt.xlabel("통화국가명", fontsize=20)
plt.ylabel("계약금액(1e7)", fontsize=20)
```

```python
#등록유형 "나라장터(G2B)"일때, 통화국가를 기준으로 계약수량 통계(2021.01.13~ 2022.01.13)
df = pd.read_csv("외자계약상세내역.csv")
df = df[df['등록유형'] == '나라장터(G2B)']
df_2 = df.loc[:,['통화국가명', '계약수량']]
grouped = df_2.groupby('통화국가명')
df_sum = grouped.sum()
df_sum = df_sum.sort_values(by ='계약수량' ,ascending=False)
df_sum.plot(kind='bar',figsize=(20,20), fontsize=15)
plt.title("나라장터 통화국가별 계약수량 통계 ", fontsize=20)
plt.xlabel("통화국가명", fontsize=20)
plt.ylabel("계약수량", fontsize=20)
```

```python
#등록유형 "나라장터(G2B)"일때, 통화국가를 기준으로 계약금액 통계(2021.01.13~ 2022.01.13)
df = pd.read_csv("외자계약상세내역.csv")
df = df[df['등록유형'] == '나라장터(G2B)']
df_2 = df.loc[:,['통화국가명', '계약금액(USD)']]
grouped = df_2.groupby('통화국가명')
df_sum = grouped.sum()
df_sum = df_sum.sort_values(by ='계약금액(USD)' ,ascending=False)
df_sum.plot(kind='bar',figsize=(20,20), fontsize=15)
plt.title("나라장터 원산국가별 계약수량 통계 ", fontsize=20)
plt.xlabel("통화국가명", fontsize=20)
plt.ylabel("계약금액(1e8)", fontsize=20)
```
